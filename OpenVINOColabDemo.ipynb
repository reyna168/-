# Camera Capture
Using a webcam to capture images for processing on the runtime.

````python
from IPython import display
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):
  js = display.Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display.display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename
````

````python
# 使用 POT 量化模型 – 以 Yolo_v4_tf 模型為例
# 執行POT之前，我們以json檔案來配置POT執行的參數
# 建立 yolo-v4-tf-pot.json

!wget https://raw.githubusercontent.com/openvinotoolkit/openvino/master/tools/pot/configs/default_quantization_spec.json
````

````python
#!cp ./default_quantization_spec.json ./yolo-v4-tf-pot.json
````

````python
#!sed -i 's/model_name/yolo-v4-tf/g' ./yolo-v4-tf-pot.json
#!sed -i 's/<MODEL_PATH>/\/content\/open_model_zoo\/models\/public\/yolo-v4-tf\/FP16\/yolo-v4-tf.xml/g' ./yolo-v4-tf-pot.json
#!sed -i 's/<PATH_TO_WEIGHTS>/\/content\/open_model_zoo\/models\/public\/yolo-v4-tf\/FP16\/yolo-v4-tf.bin/g' ./yolo-v4-tf-pot.json
````

````python
%%writefile yolo-v4-tf-pot.json
{
    "model": {
        "model_name": "yolo_v4_tf", // Model name
        "model": "/content/open_model_zoo/models/public/yolo-v4-tf/FP16/yolo-v4-tf.xml", // Path to model (.xml format)
        "weights": "/content/open_model_zoo/models/public/yolo-v4-tf/FP16/yolo-v4-tf.bin" // Path to weights (.bin format)
    },
    "engine": {
        "type": "simplified", // OR default value "type": "accuracy_checker" for non simplified mode
        "data_source": "/content/val2017" // You can specify path to directory with images. Also you can
                                        // specify template for file names to filter images to load.
                                        // Templates are unix style (This option valid only in simplified mode)
    },

    /* Optimization hyperparameters */

    "compression": {
        "algorithms": [
            {
                "name": "DefaultQuantization", // Optimization algorithm name
                "params": {
                    "preset": "performance", // A preset is a collection of optimization algorithm parameters
                                       // that will specify to the algorithm to improve which metric
                                       // the algorithm needs to concentrate. Each optimization algorithm
                                       // supports [performance, mixed, accuracy] presets

                    "stat_subset_size": 300, // Size of subset to calculate activations statistics that can be used
                                             // for quantization parameters calculation
                }
            }
        ]
    }
}
````

````python
%%writefile yolo-v4-tf-pot-Accuracy-Aware.json
{
    "model": {
        "model_name": "yolo_v4_tf", // Model name
        "model": "/content/open_model_zoo/models/public/yolo-v4-tf/FP16/yolo-v4-tf.xml", // Path to model (.xml format)
        "weights": "/content/open_model_zoo/models/public/yolo-v4-tf/FP16/yolo-v4-tf.bin" // Path to weights (.bin format)
    },
    "engine": {
        "type": "accuracy_checker", // OR default value "type": "accuracy_checker" for non simplified mode
        "config": "/content/accuracy-check.yml" 
    },

    /* Optimization hyperparameters */

    "compression": {
        "algorithms": [
            {
                "name": "AccuracyAwareQuantization", // Optimization algorithm name
                "params": {
                    "preset": "mixed", // A preset is a collection of optimization algorithm parameters
                                       // that will specify to the algorithm to improve which metric
                                       // the algorithm needs to concentrate. Each optimization algorithm
                                       // supports [performance, mixed, accuracy] presets

                    "stat_subset_size": 300, // Size of subset to calculate activations statistics that can be used
                                             // for quantization parameters calculation
                    "maximal_drop": 0.005, // Maximum accuracy drop which has to be achieved after the quantization
                }
            }
        ]
    }
}
````
